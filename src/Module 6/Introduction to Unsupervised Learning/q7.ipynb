```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
from sklearn.datasets import make_blobs
from sklearn.preprocessing import StandardScaler

# Step 1: Generate a high-dimensional dataset (e.g., 5 features, 1000 samples)
np.random.seed(42)
X, y = make_blobs(n_samples=1000, centers=4, n_features=5, random_state=42)

# Step 2: Standardize the data (important for t-SNE)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Step 3: Apply t-SNE to reduce the data to 2D
tsne = TSNE(n_components=2, random_state=42)
X_tsne = tsne.fit_transform(X_scaled)

# Step 4: Visualize the reduced data
plt.figure(figsize=(8, 6))
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=y, cmap='viridis', s=50)
plt.title("t-SNE Feature Reduction")
plt.xlabel("t-SNE Component 1")
plt.ylabel("t-SNE Component 2")
plt.colorbar(label="Cluster Label")
plt.show()

# Print the first few rows of the transformed dataset
print("\nFirst few rows of the transformed data (t-SNE reduced):")
print(pd.DataFrame(X_tsne, columns=["Component 1", "Component 2"]).head())
```
