```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

# Generate a simple dataset with 2 features and 3 clusters
X, _ = make_blobs(n_samples=300, centers=3, random_state=42)

# Visualize the generated dataset
plt.scatter(X[:, 0], X[:, 1], c='black', marker='o')
plt.title("Generated Data for Clustering")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()

# Apply K-Means clustering
kmeans = KMeans(n_clusters=3, random_state=42)  # Specify 3 clusters
kmeans.fit(X)

# Get the cluster centers and labels
cluster_centers = kmeans.cluster_centers_
labels = kmeans.labels_

# Visualize the clustering result
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', marker='o')  # Points colored by cluster
plt.scatter(cluster_centers[:, 0], cluster_centers[:, 1], s=200, c='red', marker='X', label='Centroids')  # Cluster centers
plt.title("Clustering Result with K-Means")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()
```
